```python
# Import required libraries
import pandas as pd

# Initialize variables
fInputHawaiiStations='clean_hawaii_stations.csv'
fInputHawaiiMeasurements='clean_hawaii_measurements.csv'

# Open input file, read into frame
dfInputHawaiiStations = pd.read_csv(fInputHawaiiStations)
dfInputHawaiiMeasurements = pd.read_csv(fInputHawaiiMeasurements)
```

```python
# Drop index columns
dfInputHawaiiStations.drop('Unnamed: 0', axis=1, inplace=True)
dfInputHawaiiMeasurements.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Import libraries

from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
from sqlalchemy import Column, Integer, String, Float, Date
import sqlite3

# Create classes
class Stations(Base):
    __tablename__ = 'stations'
    station = Column(String(255), primary_key=True)
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
    
class Measurements(Base):
    __tablename__ = 'measurements'
    index = Column(Integer, primary_key=True)
    station = Column(String(255))
    date = Column(Date)
    prcp = Column(Float)
    tobs = Column(Integer)

# Create Database Connection
# ----------------------------------
engine = create_engine('sqlite:///hawaii.sqlite')
conn = engine.connect()

# Create a "Metadata" Layer That Abstracts our SQL Database
# ----------------------------------
Base.metadata.create_all(engine)

# Write dataframes to tables
dfInputHawaiiStations.to_sql('stations', conn, schema=None, index=False, if_exists='append', chunksize=None, dtype=None)
dfInputHawaiiMeasurements.to_sql('measurements', conn, schema=None, index=True, if_exists='append', chunksize=None, dtype=None)
```
