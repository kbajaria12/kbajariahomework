```python
import os
import csv

# Define static variables
fileData1Name='employee_data2.csv'
colInputEmpID=0;
colInputName=1;
colInputDate=2;
colInputSSN=3;
colInputState=4;

dictStates = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY',
}

# Open file as csv and load into a list
fileData1=open(fileData1Name)
readerData1=csv.reader(fileData1, delimiter=",")
listData1=list(readerData1)

# Print headings
print ('Emp ID,First Name,LastName,DOB,SSN,State')

# Initialize accumulators and counters
RowCount=0
CurrentRow=0

# Determine total rows of data
RowCount=len(listData1)

# Month count is the total rows minus the header row
EmpCount=RowCount-1

# Set up file iterator
for CurrentRow in range(1,RowCount):
    
    # Reformat output fields
    
    # Employee ID
    OutputEmpID=listData1[CurrentRow][colInputEmpID]
    InputName=listData1[CurrentRow][colInputName]
    
    # Employee Name
    SpaceNameIndex=InputName.find(" ")
    EndOfFirstName=SpaceNameIndex
    StartOfLastName=SpaceNameIndex + 1
    OutputFirstName=InputName[0:EndOfFirstName]
    OutputLastName=InputName[StartOfLastName:len(InputName)]
    
    # Date
    InputDate=listData1[CurrentRow][colInputDate]
    InputYear=InputDate[0:4]
    InputMonth=InputDate[5:7]
    InputDay=InputDate[8:10]
    OutputDate=InputDay+'/'+InputMonth+'/'+InputYear
    
    # SSN
    InputSSN=listData1[CurrentRow][colInputSSN]
    OutputSSN='***-**-'+InputSSN[7:11]
    
    # State
    InputState=listData1[CurrentRow][colInputState]
    OutputState=dictStates.get(InputState)
    
    # Output row
    OutputRow=OutputEmpID+","+OutputFirstName+","+OutputLastName+","+OutputDate+","+OutputSSN+","+OutputState
    print(OutputRow)
    
# End file iterator
```
